## 2. Introduction


**Project Lab 2** examines a global earthquake record and is divided into two parts.

The data file for this lab contains earthquake data for earthquakes with magnitudes 4 or above ($M \geq 4$) that have occurred globally in the 21st century, i.e. between `2000-01-01` and `2022-11-24`.
Earthquake data for this lab were retrieved from **IRIS**: [&nearr; Incorporated Research Institutions for Seismology](http://ds.iris.edu/ieb/index.html?format=text&nodata=404&starttime=1970-01-01&endtime=2025-01-01&minmag=0&maxmag=10&mindepth=0&maxdepth=900&orderby=time-desc&src=&limit=1000&maxlat=80.56&minlat=-80.56&maxlon=180.00&minlon=-180.00&zm=2&mt=ter). 

**Part 1:** Do some basic exploratory data analysis of this data set to look at the overall properties of global seismicity (earthquakes).

**Part 2:** Further explore some aspects of seismicity via qualitative observations of figures and quantitative modeling.

Read the background reading (`project_lab2_background.ipynb`) BEFORE attempting the lab.  This contains information that will be applicable to different parts of the project lab (data files and formats, distances on a sphere etc).  In the project lab writeups we refer you back to sections in this reading where needed / appropriate.

You will also need to have gone over the notebook `pl2_pandas_getfamiliar.ipynb` which contains an overview of loading and manipulating a similar dataset to the one used in this project lab. Some of this may be repeated in a wk13 class notebook.
