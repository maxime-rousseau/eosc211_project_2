```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import earthquake_fns as quake   # our module

# ---- File paths (relative to the project_lab2 folder) ----
coasts_file = "m_coasts.csv"
plates_file = "all_boundaries.csv"
quakes_file = "merged_eq_010100_112025_mag4.csv"

# ---- Use the functions from earthquake_fns ----
lon_coast, lat_coast = quake.get_coastlines(coasts_file)
pb_dict = quake.get_plate_boundaries(plates_file)

df_quakes = quake.get_earthquakes(quakes_file)
lats, lons, depths, magnitudes, times = quake.parse_earthquakes_to_np(df_quakes)
```

```python
# ----- Largest-magnitude earthquake -----
imax_mag = np.argmax(magnitudes)

largest_quake = {
    "magnitude": float(magnitudes[imax_mag]),
    "time":      times[imax_mag],
    "latitude":  float(lats[imax_mag]),
    "longitude": float(lons[imax_mag]),
    "depth_km":  float(depths[imax_mag])
}

print(
    f"Largest earthquake (2000–2025): "
    f"M {largest_quake['magnitude']:.1f}, "
    f"time {largest_quake['time']:%Y-%m-%d %H:%M}, "
    f"lat {largest_quake['latitude']:.2f}°, "
    f"lon {largest_quake['longitude']:.2f}°, "
    f"depth {largest_quake['depth_km']:.1f} km"
)

# ----- Deepest earthquake (max depth) -----
imax_depth = np.argmax(depths)

deepest_quake = {
    "magnitude": float(magnitudes[imax_depth]),
    "time":      times[imax_depth],
    "latitude":  float(lats[imax_depth]),
    "longitude": float(lons[imax_depth]),
    "depth_km":  float(depths[imax_depth])
}

print(
    f"Deepest earthquake (2000–2025): "
    f"M {deepest_quake['magnitude']:.1f}, "
    f"time {deepest_quake['time']:%Y-%m-%d %H:%M}, "
    f"lat {deepest_quake['latitude']:.2f}°, "
    f"lon {deepest_quake['longitude']:.2f}°, "
    f"depth {deepest_quake['depth_km']:.1f} km"
)
```
